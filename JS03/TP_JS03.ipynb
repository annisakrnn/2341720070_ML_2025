# Tugas Praktikum

Deskripsi Tugas

Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
2. Lakukan proses encoding pada kolom "diagnosis".
3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.
4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Import library yang diperlukan
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 1. Memuat dataset wbc
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['diagnosis'] = data.target  # Kolom target (0 untuk Benign, 1 untuk Malignant)

# Menambahkan kolom 'id' sebagai dummy untuk tugas
df['id'] = range(1, len(df) + 1)

# Pemisahan variabel yang dapat digunakan dan tidak dapat digunakan
# Kolom 'id' dianggap tidak dapat digunakan
kolom_digunakan = data.feature_names.tolist() + ['diagnosis']
kolom_tidak_digunakan = ['id']
print("Variabel yang dapat digunakan:", kolom_digunakan)
print("Variabel yang tidak dapat digunakan:", kolom_tidak_digunakan)

# Membuat dataframe hanya dengan kolom yang dapat digunakan
df_digunakan = df[kolom_digunakan]

# 2. Encoding kolom 'diagnosis'
# Dataset asli sudah bernilai 0 (Benign) dan 1 (Malignant), tetapi kita ubah ke 'B' dan 'M' lalu encode ulang
df_digunakan['diagnosis'] = df_digunakan['diagnosis'].map({0: 'B', 1: 'M'})
df_digunakan['diagnosis'] = df_digunakan['diagnosis'].map({'B': 0, 'M': 1})
print("\nHasil encoding kolom 'diagnosis' (5 baris pertama):")
print(df_digunakan['diagnosis'].head())

# 3. Standarisasi kolom numerik
scaler = StandardScaler()
kolom_numerik = data.feature_names  # Semua fitur adalah numerik
df_digunakan[kolom_numerik] = scaler.fit_transform(df_digunakan[kolom_numerik])
print("\nDataframe setelah standarisasi (5 baris pertama):")
print(df_digunakan.head())

# 4. Stratified split data (80:20)
X = df_digunakan.drop('diagnosis', axis=1)  # Fitur
y = df_digunakan['diagnosis']  # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Cek ukuran data
print("\nUkuran data latih:", X_train.shape)
print("Ukuran data uji:", X_test.shape)

# Cek distribusi kelas di data latih dan uji
print("\nDistribusi kelas di data latih:")
print(y_train.value_counts(normalize=True))
print("\nDistribusi kelas di data uji:")
print(y_test.value_counts(normalize=True))
```
